```python
import matplotlib.pyplot as plt
import numpy as np

from neo_detect.phase_fnc import Phase_Function
```

```python
x = np.linspace(0, 180, 180)
y = np.zeros_like(x)
for i in range(len(x)):
    y[i] = Phase_Function(x[i], G = 0.15).phase_function()
```

```python
plt.plot(x, y)
plt.xlabel('Phase Angles (degrees)')
plt.ylabel('Phase Function Values')
plt.title('Phase Function Values vs Phase Angles')
plt.savefig('phase_function_plot.png')
```
